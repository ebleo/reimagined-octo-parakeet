````python
s = '<tr align="right"><td>3</td><td>Matthew</td><td>Brittany</td>'
````

````python
match = re.search(r'>([\w\d]+)<', s)
print('found', match.group()) # this will print the entire matching string
print('our match group is: ', match.group(1)) # this prints our match group
````

The following is a function which extracts the names as per the specifications of the Google babynames excercise.

Assuming that the `re` module has been imported.

A few notes on Google's solution:
- their solution is much more elegant, instead of pulling out names and numbers separately...
  - they first unpack all the desired strings in tuples of (number, boyname, girlname)
  - in a loop through a list of tuples, each name is stuck with a number and put into a master list
  
```python
  names_to_rank =  {}
  for rank_tuple in tuples:
    (rank, boyname, girlname) = rank_tuple  # unpack the tuple into 3 vars
    if boyname not in names_to_rank:
      names_to_rank[boyname] = rank
    if girlname not in names_to_rank:
      names_to_rank[girlname] = rank
      ```

````python
def extract_names(filename):
    file = open(filename, 'r')
    tx = file.read()
    file.close()
    
    matchf = re.findall(r'd>([A-Za-z]+)<', tx)

    m = []
    f = []
    for i in range(len(matchf)):
        if i%2 > 0: # odd
            f.append(matchf[i])
        else:
            m.append(matchf[i])

    a = len(m)
    b = len(f)
    x = max(a,b)

    matchd = re.findall(r'>(\d+)<', tx)

    m_asint = []

    for i in range(x):
        m_asint.append(int(matchd[i]))

    d = {}
    for i in range(x):
        d[m_asint[i]] = [m[i], f[i]] 

    year = re.search(r'Popularity in (\d{4})', tx)
    year_str = year.group(1)

    babylist = []
    babylist.append(year_str)

    rankf = []
    rankm = []

    for i in range(x):
        rankf.append(f[i] + ' ' + str(matchd[i]))
        rankm.append(m[i] + ' ' + str(matchd[i]))

    rankall = rankf + rankm
    rankall = sorted(rankall)

    babylist += rankall
    
#     following code returns list when not in console
#     for item in babylist:
#         print(item)
    
    return babylist
````
